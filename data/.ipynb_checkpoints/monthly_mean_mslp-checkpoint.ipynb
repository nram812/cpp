```python
import xarray as xr
from dask.diagnostics import ProgressBar
import matplotlib.pyplot as plt
```

```python
mslp_data = '/nesi/project/niwa00004/ML_DATA/ERA5/u_v_w_experiments/'
# loading the downloaded dataset
complete_dset = xr.open_dataset(f"{mslp_data}mean_sea_level_pressure.nc", chunks={"time":1000})
with ProgressBar():
    monthly_means = complete_dset.resample(time = '1MS').mean().compute()
```

```python
version1 = monthly_means['msl'].isel(expver =0).sel(time = slice(None, "2021-04-25"))
version2 = monthly_means['msl'].isel(expver =1).sel(time = slice("2021-05-01", None))
```

```python
merged = xr.concat([version1, version2], dim ="time")
```

```python
fig, ax = plt.subplots()
version1.isel(latitude =100,
                                          longitude = 100).sel(time = slice("2021", "2021-04-01")).plot(ax = ax,
                                                                                                        marker ='o')
version2.isel(latitude =100,
                                          longitude = 100).plot(ax = ax)
merged.isel(latitude =100,
                                          longitude = 100).sel(time = slice("2021", "2021")).plot(ax = ax, color ='r')
fig.show()
```

```python
merged.to_netcdf('/nesi/project/niwa00004/rampaln/CAOA2101/cpp-indices/data/monthly_mean_mslp.nc')
```
